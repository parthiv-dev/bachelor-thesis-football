# Multiplying scores by minutesPlayed be able to calculate weighted average

```python
# Multiplying all columns starting with 'performance' or 'profile' by minutesPlayed for df_cb, df_fb, df_mf, df_st
for df in [df_cb, df_fb, df_mf, df_w, df_st]:
    for col in df.columns:
        if col.startswith('performance') or col.startswith('profile'):
            df[col] = df[col] * df['minutesPlayed']

df_cb.head()
```

# Creating new dataframes where we group stats by player

```python
# Average profile and performance scores for each player in a dataframe
# Sum of number of gameweeks, minutes played

# Groupby playerId, playerName, teamName, position and aggregate the data: profilescore_cb_def	profilescore_cb_off	performance_cb_def	performance_cb_off	performance_nonprofile	performancescore

df_cb_comb = df_cb.groupby(by=['playerId', 'playerName', 'teamName', 'position']).agg({'gameweek': 'count', 
                                                                                        'minutesPlayed': 'sum', 
                                                                                        'profilescore_cb_def': 'mean', 
                                                                                        'profilescore_cb_off': 'mean', 
                                                                                        'performance_cb_def': 'mean', 
                                                                                        'performance_cb_def': 'mean', 
                                                                                        'performance_nonprofile': 'mean', 
                                                                                        'performancescore': 'mean'}).reset_index()

# Groupby playerId, playerName, teamName and aggregate the data: profilescore_fb_def	profilescore_fb_off	performance_fb_def	performance_fb_off	performance_nonprofile	performancescore

df_fb_comb = df_fb.groupby(by=['playerId', 'playerName', 'teamName', 'position']).agg({'gameweek': 'count',
                                                                                        'minutesPlayed': 'sum', 
                                                                                        'profilescore_fb_def': 'mean', 
                                                                                        'profilescore_fb_off': 'mean', 
                                                                                        'performance_fb_def': 'mean', 
                                                                                        'performance_fb_off': 'mean', 
                                                                                        'performance_nonprofile': 'mean', 
                                                                                        'performancescore': 'mean'}).reset_index()

# Groupby playerId, playerName, teamName and aggregate the data: profilescore_mf_playmaker	profilescore_mf_creator	profilescore_mf_def	profilescore_mf_off	performance_mf_def	performance_mf_playmaker	performance_mf_creator	performance_mf_off	performance_nonprofile	performancescore

df_mf_comb = df_mf.groupby(by=['playerId', 'playerName', 'teamName', 'position']).agg({'gameweek': 'count',
                                                                                        'minutesPlayed': 'sum', 
                                                                                        'profilescore_mf_playmaker': 'mean', 
                                                                                        'profilescore_mf_creator': 'mean', 
                                                                                        'profilescore_mf_def': 'mean', 
                                                                                        'profilescore_mf_off': 'mean', 
                                                                                        'performance_mf_def': 'mean', 
                                                                                        'performance_mf_playmaker': 'mean', 
                                                                                        'performance_mf_creator': 'mean', 
                                                                                        'performance_mf_off': 'mean', 
                                                                                        'performance_nonprofile': 'mean', 
                                                                                        'performancescore': 'mean'}).reset_index()


# Groupby playerId, playerName, teamName and aggregate the data: profilescore_wing_creator	profilescore_wing_dribbler	profilescore_wing_runner	performance_wing_creator	performance_wing_dribbler	performance_wing_runner	performance_nonprofile	performancescore

df_w_comb = df_w.groupby(by=['playerId', 'playerName', 'teamName', 'position']).agg({'gameweek': 'count',
                                                                                        'minutesPlayed': 'sum', 
                                                                                        'profilescore_wing_creator': 'mean', 
                                                                                        'profilescore_wing_dribbler': 'mean', 
                                                                                        'profilescore_wing_runner': 'mean', 
                                                                                        'performance_wing_creator': 'mean', 
                                                                                        'performance_wing_dribbler': 'mean', 
                                                                                        'performance_wing_runner': 'mean', 
                                                                                        'performance_nonprofile': 'mean', 
                                                                                        'performancescore': 'mean'}).reset_index()

# Groupby playerId, playerName, teamName and aggregate the data: profilescore_striker_finisher	profilescore_striker_creator	profilescore_striker_dribbler	performance_striker_finisher	performance_striker_creator	performance_striker_dribbler	performance_nonprofile	performancescore

df_st_comb = df_st.groupby(by=['playerId', 'playerName', 'teamName', 'position']).agg({'gameweek': 'count',
                                                                                        'minutesPlayed': 'sum', 
                                                                                        'profilescore_striker_finisher': 'mean', 
                                                                                        'profilescore_striker_creator': 'mean', 
                                                                                        'profilescore_striker_dribbler': 'mean', 
                                                                                        'performance_striker_finisher': 'mean', 
                                                                                        'performance_striker_creator': 'mean', 
                                                                                        'performance_striker_dribbler': 'mean', 
                                                                                        'performance_nonprofile': 'mean', 
                                                                                        'performancescore': 'mean'}).reset_index()
df_cb_comb
```

```python

for df in [df_cb_comb, df_fb_comb, df_mf_comb, df_w_comb, df_st_comb]:
    # Players with less than 500 minutes played are removed
    df.drop(df[df['minutesPlayed'] < 500].index, inplace=True)

    # Any column starting with 'performance' or 'profile' is divided by minutesPlayed
    for col in df.columns:
        if col.startswith('performance') or col.startswith('profile'):
            df[col] = df[col] / df['minutesPlayed']

        # Round all columns to 2 decimal places
        df[col] = df[col].round(2)
    df.rename(columns={'gameweek': 'number_gameweeks'}, inplace=True)

    # Sorting dataframe by minutesPlayed in descending order
    df.sort_values(by='performance_nonprofile', ascending=False, inplace=True)

df_cb_comb
```
