```python
# Imports

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

import os
import datetime as dt
```

```python
df = pd.DataFrame()

# Create synthetic data for 4 columns (of size 10) with mean 10 and std 1,2,4,3

std = [1, 2, 4, 3]

for i in range(4):
    df[f'metric_{i}'] = np.random.normal(10, std[i], 10)

df
```

```python
# Plot data using matplotlib with each row as a line
plt.figure(figsize=(10, 7))
plt.plot(df.T, marker='o')
plt.title('Synthetic players')
plt.xlabel('Performance Metric')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Save df as csv file with name of current file and current date and time
# pathname = os.path.basename(__file__)
pathname = "synthetic_data"
current_time = dt.datetime.now().strftime("%Y%m%d_%H%M%S")
csv_filename = f"data_output/data_{pathname}_{current_time}.csv"
df.to_csv(csv_filename, index=False)
```
